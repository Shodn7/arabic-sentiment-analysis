```python
import pandas as pd

# رفع الملف
from google.colab import files
uploaded = files.upload()

# قراءة البيانات
df = pd.read_csv(list(uploaded.keys())[0])

# قائمة كلمات إيجابية وسلبية بسيطة
positive_words = ['ممتاز', 'جميل', 'سهل', 'سريعة', 'شكراً', 'رائع', 'مميزة', 'أفضل', 'يسهل', 'ناجح', 'فوري']
negative_words = ['سيء', 'صعب', 'يعلق', 'مشاكل', 'بطئ', 'تأخير', 'مزعج', 'ما يشتغل', 'خلل', 'أسوأ', 'خرب']

# دالة التصنيف
def classify_sentiment(text):
    text = str(text)
    if any(word in text for word in positive_words):
        return 'إيجابي'
    elif any(word in text for word in negative_words):
        return 'سلبي'
    else:
        return 'محايد'

# تطبيق التصنيف
df['sentiment'] = df['text'].apply(classify_sentiment)
df[['text', 'sentiment']].head()
```

```python
import matplotlib.pyplot as plt

df['sentiment'].value_counts().plot.pie(
    autopct='%1.1f%%',
    startangle=90,
    shadow=True,
    colors=['#5cb85c', '#d9534f', '#f0ad4e']
)
plt.title('تحليل مشاعر التغريدات عن توكلنا')
plt.ylabel('')
plt.show()
```

```python
import matplotlib.pyplot as plt
import arabic_reshaper
from bidi.algorithm import get_display

# إعادة تشكيل الكلمات بالعربي (محايد، سلبي، إيجابي)
labels = df['sentiment'].value_counts().index.tolist()
reshaped_labels = [arabic_reshaper.reshape(label) for label in labels]
bidi_labels = [get_display(label) for label in reshaped_labels]

# الرسم البياني
df['sentiment'].value_counts().plot.pie(
    autopct='%1.1f%%',
    startangle=90,
    shadow=True,
    labels=bidi_labels,
    colors=['#5cb85c', '#d9534f', '#f0ad4e']
)
plt.title(get_display(arabic_reshaper.reshape('تحليل مشاعر التغريدات عن توكلنا')))
plt.ylabel('')
plt.show()
```
