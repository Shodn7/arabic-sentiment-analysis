```python
import pandas as pd

# ÿ±ŸÅÿπ ÿßŸÑŸÖŸÑŸÅ
from google.colab import files
uploaded = files.upload()

# ŸÇÿ±ÿßÿ°ÿ© ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
df = pd.read_csv(list(uploaded.keys())[0])

# ŸÇÿßÿ¶ŸÖÿ© ŸÉŸÑŸÖÿßÿ™ ÿ•Ÿäÿ¨ÿßÿ®Ÿäÿ© Ÿàÿ≥ŸÑÿ®Ÿäÿ© ÿ®ÿ≥Ÿäÿ∑ÿ©
positive_words = ['ŸÖŸÖÿ™ÿßÿ≤', 'ÿ¨ŸÖŸäŸÑ', 'ÿ≥ŸáŸÑ', 'ÿ≥ÿ±Ÿäÿπÿ©', 'ÿ¥ŸÉÿ±ÿßŸã', 'ÿ±ÿßÿ¶ÿπ', 'ŸÖŸÖŸäÿ≤ÿ©', 'ÿ£ŸÅÿ∂ŸÑ', 'Ÿäÿ≥ŸáŸÑ', 'ŸÜÿßÿ¨ÿ≠', 'ŸÅŸàÿ±Ÿä']
negative_words = ['ÿ≥Ÿäÿ°', 'ÿµÿπÿ®', 'ŸäÿπŸÑŸÇ', 'ŸÖÿ¥ÿßŸÉŸÑ', 'ÿ®ÿ∑ÿ¶', 'ÿ™ÿ£ÿÆŸäÿ±', 'ŸÖÿ≤ÿπÿ¨', 'ŸÖÿß Ÿäÿ¥ÿ™ÿ∫ŸÑ', 'ÿÆŸÑŸÑ', 'ÿ£ÿ≥Ÿàÿ£', 'ÿÆÿ±ÿ®']

# ÿØÿßŸÑÿ© ÿßŸÑÿ™ÿµŸÜŸäŸÅ
def classify_sentiment(text):
    text = str(text)
    if any(word in text for word in positive_words):
        return 'ÿ•Ÿäÿ¨ÿßÿ®Ÿä'
    elif any(word in text for word in negative_words):
        return 'ÿ≥ŸÑÿ®Ÿä'
    else:
        return 'ŸÖÿ≠ÿßŸäÿØ'

# ÿ™ÿ∑ÿ®ŸäŸÇ ÿßŸÑÿ™ÿµŸÜŸäŸÅ
df['sentiment'] = df['text'].apply(classify_sentiment)
df[['text', 'sentiment']].head()
```

```python
import matplotlib.pyplot as plt

df['sentiment'].value_counts().plot.pie(
    autopct='%1.1f%%',
    startangle=90,
    shadow=True,
    colors=['#5cb85c', '#d9534f', '#f0ad4e']
)
plt.title('ÿ™ÿ≠ŸÑŸäŸÑ ŸÖÿ¥ÿßÿπÿ± ÿßŸÑÿ™ÿ∫ÿ±ŸäÿØÿßÿ™ ÿπŸÜ ÿ™ŸàŸÉŸÑŸÜÿß')
plt.ylabel('')
plt.show()
```
